
#### 2] . Write a Pandas  program to import excel data (employee.xlsx ) into a Pandas dataframe and s to sort the records by the hire date column.

```python
#import pandas as pd
import pandas as pd

#import NumPy as np
import numpy as np

# making data frame from xlsx file
df = pd.read_excel('D:\employee1.xlsx')

result = df.sort_values('hire_date')

result
```

 
#### 3] . Write a Pandas program to import excel data (employee.xlsx ) into a Pandas dataframe and find a list of employees of a specified year.
